```python
import pickle as pkl
path = 'flock_feat_name.pkl'
```

```python
with open(path, 'rb') as f:
    feat_names = pkl.load(f)
```

```python
feat_names = [f.replace('.', '_') for f in feat_names]
feat_names
```

```python
# make the first letter of each word uppercase
feat_names = [''.join([w[0].upper() + w[1:] for w in f.split('_')]) for f in feat_names]
feat_names
```

```python
# write functions
for i, f in enumerate(feat_names):
    print("    def get{}FeatureValue(self):\n        return self.features[{}]\n".format(f, i))
```
